```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Arrivingvehicle=ctrl.Antecedent(np.arange(0,101,1),'Arrivingvehicle')
Arrivingvehicle['less']=fuzz.trimf(Arrivingvehicle.universe,[0,10,20])
Arrivingvehicle['medium']=fuzz.trimf(Arrivingvehicle.universe,[14,33,50])
Arrivingvehicle['high']=fuzz.trimf(Arrivingvehicle.universe,[40,70,101])
Arrivingvehicle.view()
```

```python
Queuingvehicle=ctrl.Antecedent(np.arange(0,101,1),'Queuingvehicle')
Queuingvehicle['less']=fuzz.trimf(Queuingvehicle.universe,[0,10,20])
Queuingvehicle['medium']=fuzz.trimf(Queuingvehicle.universe,[14,33,50])
Queuingvehicle['high']=fuzz.trimf(Queuingvehicle.universe,[40,70,101])
Queuingvehicle.view()
```

```python
Fog=ctrl.Antecedent(np.arange(50,501,1),'Fog')
Fog['less']=fuzz.trimf(Fog.universe,[330,415,500])
Fog['medium']=fuzz.trimf(Fog.universe,[150,225,350])
Fog['high']=fuzz.trimf(Fog.universe,[50,110,170])
Fog.view()
```

```python
GLduration=ctrl.Consequent(np.arange(0,61,1),'GLduration')
GLduration.automf(names=['short','medium','long'])
GLduration.view()
```

```python
df=pd.DataFrame(
    {
        'Arriving Vehicle':np.random.randint(1,101,150),
        'Queueing Vehicle':np.random.randint(1,101,150),
        'Fog':np.random.randint(50,501,150)
    }
)
data=df.values
data
```

```python
#Generating Target Column using fuzzy logic.

rule1=ctrl.Rule(Arrivingvehicle['high'] & Queuingvehicle['less'] & Fog['less'], GLduration['short'])
rule2=ctrl.Rule(Arrivingvehicle['high'] & Queuingvehicle['less'] & Fog['medium'], GLduration['medium'])
rule3=ctrl.Rule(Arrivingvehicle['high'] & Queuingvehicle['less'] & Fog['high'], GLduration['long'])
rule4=ctrl.Rule(Arrivingvehicle['high'] & Queuingvehicle['medium'] & Fog['less'], GLduration['medium'])
rule5=ctrl.Rule(Arrivingvehicle['high'] & Queuingvehicle['medium'] & Fog['medium'], GLduration['medium'])
rule6=ctrl.Rule(Arrivingvehicle['high'] & Queuingvehicle['medium'] & Fog['high'], GLduration['long'])
rule7=ctrl.Rule(Arrivingvehicle['high'] & Queuingvehicle['high'] & Fog['less'], GLduration['long'])
rule8=ctrl.Rule(Arrivingvehicle['high'] & Queuingvehicle['high'] & Fog['medium'], GLduration['long'])
rule9=ctrl.Rule(Arrivingvehicle['high'] & Queuingvehicle['high'] & Fog['high'], GLduration['long'])
rule10=ctrl.Rule(Arrivingvehicle['medium'] & Queuingvehicle['less'] & Fog['less'], GLduration['short'])
rule11=ctrl.Rule(Arrivingvehicle['medium'] & Queuingvehicle['less'] & Fog['medium'], GLduration['medium'])
rule12=ctrl.Rule(Arrivingvehicle['medium'] & Queuingvehicle['less'] & Fog['high'], GLduration['medium'])
rule13=ctrl.Rule(Arrivingvehicle['medium'] & Queuingvehicle['medium'] & Fog['less'], GLduration['medium'])
rule14=ctrl.Rule(Arrivingvehicle['medium'] & Queuingvehicle['medium'] & Fog['medium'], GLduration['medium'])
rule15=ctrl.Rule(Arrivingvehicle['medium'] & Queuingvehicle['medium'] & Fog['high'], GLduration['medium'])
rule16=ctrl.Rule(Arrivingvehicle['medium'] & Queuingvehicle['high'] & Fog['less'], GLduration['long'])
rule17=ctrl.Rule(Arrivingvehicle['medium'] & Queuingvehicle['high'] & Fog['medium'], GLduration['long'])
rule18=ctrl.Rule(Arrivingvehicle['medium'] & Queuingvehicle['high'] & Fog['high'], GLduration['medium'])
rule19=ctrl.Rule(Arrivingvehicle['less'] & Queuingvehicle['less'] & Fog['less'], GLduration['short'])
rule20=ctrl.Rule(Arrivingvehicle['less'] & Queuingvehicle['less'] & Fog['medium'], GLduration['medium'])
rule21=ctrl.Rule(Arrivingvehicle['less'] & Queuingvehicle['less'] & Fog['high'], GLduration['medium'])
rule22=ctrl.Rule(Arrivingvehicle['less'] & Queuingvehicle['medium'] & Fog['less'], GLduration['medium'])
rule23=ctrl.Rule(Arrivingvehicle['less'] & Queuingvehicle['medium'] & Fog['medium'], GLduration['long'])
rule24=ctrl.Rule(Arrivingvehicle['less'] & Queuingvehicle['medium'] & Fog['high'], GLduration['medium'])
rule25=ctrl.Rule(Arrivingvehicle['less'] & Queuingvehicle['high'] & Fog['less'], GLduration['short'])
rule26=ctrl.Rule(Arrivingvehicle['less'] & Queuingvehicle['high'] & Fog['medium'], GLduration['medium'])
rule27=ctrl.Rule(Arrivingvehicle['less'] & Queuingvehicle['high'] & Fog['high'], GLduration['long'])
rules=[rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,
       rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,
       rule23,rule24,rule25,rule26,rule27]
CS=ctrl.ControlSystem(rules)
GLD=ctrl.ControlSystemSimulation(CS)
li=[]
for i in range(0,150):
    GLD.input['Arrivingvehicle']=data[i,0]
    GLD.input['Queuingvehicle']=data[i,1]
    GLD.input['Fog']=data[i,2]
    GLD.compute()
    li.append(GLD.output['GLduration'])
```

```python
df['GSTiming']=li   #Adding Target column to dataset.
```

```python
df=df.drop_duplicates(subset='GSTiming',keep='first',inplace=False)
```
